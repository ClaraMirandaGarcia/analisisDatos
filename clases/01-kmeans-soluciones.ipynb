# K-Means con Scikit-learn

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
```

```python
# TODO: Importa la clase KMeans de sklearn
from sklearn.cluster import KMeans

# TODO: Inicializar y entrenar el modelo KMeans con 3 clusters
kmeans = KMeans(n_clusters=3, random_state=1234)
kmeans.fit(X)
```

```python
# TODO: Obtener a que cluster pertenece cada dato
y_kmeans = kmeans.predict(X)
```

```python
import matplotlib.pyplot as plt

# TODO: Visualiza los clusters para las primeras dos dimensiones y sus centroides. Sigue los siguientes pasos:

# TODO: Paso 1: Obtener los centroides
# TIP: Mira los parámetros de kmeans
centers = kmeans.cluster_centers_

# TODO: Paso 2: Pinta un scatter de las dos primeras dimensiones de X (valores reales)
# y dale un color según su predicción
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap="viridis")

# TODO: Paso 3: Pinta otro scatter con los centroides.
# TIP: Para mejorar la visualiación pon un alpha (transparencia) de 0.5
plt.scatter(centers[:, 0], centers[:, 1], c="red", s=200, alpha=0.5)

plt.title("Clustering de Iris con K-Means y sus centroides")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

# Obtener el número óptimo de clusters para K-Means

```python
iris = load_iris()
X = iris.data
```

```python
inertias = []

for k in range(1, 11):
  # TODO: Instancia un KMeans que se ejecute cada vez con el número de clusters correspondiente.
  # TIP: Para evitar el warning, inicializa el parámetro n_init por ejemplo a 10.
  kmeans = KMeans(n_clusters=k, n_init=10, random_state=1234)
  kmeans.fit(X)
  # TODO: Añade a la lista "inertias" la inercia en cada iteración.
  # TIP: La inercia es la suma de las distancias al cuadrado y KMeans la calcula
  inertias.append(kmeans.inertia_)
```

```python
# TODO: Imprime el gráfico para visualizar dónde está el número óptimo de clusters
plt.plot(range(1, 11), inertias, "-o")
plt.title("Elbow Method")
plt.xlabel("Número de clusters (k)")
plt.ylabel("Suma de distancias cuadráticas (Inercia)")
plt.xticks(range(1, 11))
plt.show()
```
