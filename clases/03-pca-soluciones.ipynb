# Reducci贸n de dimensionalidad con PCA

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target
```

```python
# TODO: Importa la clase PCA de sklearn
from sklearn.decomposition import PCA

# TODO: Inicializar un PCA para reducir a dimensi贸n 2
pca = PCA(n_components=2)

# TODO: Obtener los datos con la dimensi贸n reducida
X_r = pca.fit_transform(X)
```

```python
import matplotlib.pyplot as plt

plt.figure()
colors = ["navy", "turquoise", "darkorange"]
lw = 2


for color, i, target_name in zip(colors, [0, 1, 2], iris.target_names):
  # TODO: Haz un scatter plot con las dos componentes de X_r y en cada
  # iteraci贸n imprime la clase correspondiente con el color que
  # corresponda de "colors".
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("PCA de IRIS dataset")
plt.show()
```
